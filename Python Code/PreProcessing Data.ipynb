# Import Libraries and Datasets

```python
import pandas as pd, numpy as np
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

!pip install nltk
nltk.download('punkt_tab')
nltk.download('punkt')

pd.set_option('display.max_columns', None)
```

```python
avengers = pd.read_csv("Raw Datasets\Avengers Endgame\movieReviews.csv")
Forrest_Gump = pd.read_csv('Raw Datasets\Forrest Gump\movieReviews.csv')
John_Wick = pd.read_csv("Raw Datasets\John Wick Chapter 3  Parabellum\movieReviews.csv")
Joker = pd.read_csv("Raw Datasets\Joker\movieReviews.csv")
Morbis = pd.read_csv("Raw Datasets\Morbius\movieReviews.csv")
Pulp_Fiction = pd.read_csv("Raw Datasets\Pulp Fiction\movieReviews.csv")
SpiderMan = pd.read_csv("Raw Datasets\SpiderMan No Way Home\movieReviews.csv")
The_Avengers = pd.read_csv("Raw Datasets\The Avengers\movieReviews.csv")
The_dark_night = pd.read_csv("Raw Datasets\The Dark Knight\movieReviews.csv")
Thor_Ragnarok = pd.read_csv("Raw Datasets\Thor Ragnarok\movieReviews.csv")
```

# Data Preprocessing

## Create movieID for each movie

```python
ID = [1] * avengers.shape[0]
avengers['MovieID'] = ID

ID = [2] * Forrest_Gump.shape[0]
Forrest_Gump['MovieID'] = ID

ID = [3] * John_Wick.shape[0]
John_Wick['MovieID'] = ID

ID = [4] * Joker.shape[0]
Joker['MovieID'] = ID

ID = [5] * Morbis.shape[0]
Morbis['MovieID'] = ID

ID = [6] * Pulp_Fiction.shape[0]
Pulp_Fiction['MovieID'] = ID

ID = [7] * SpiderMan.shape[0]
SpiderMan['MovieID'] = ID

ID = [8] * The_Avengers.shape[0]
The_Avengers['MovieID'] = ID

ID = [9] * The_dark_night.shape[0]
The_dark_night['MovieID'] = ID

ID = [10] * Thor_Ragnarok.shape[0]
Thor_Ragnarok['MovieID'] = ID
```

## Combining datasets

```python
dfs = [avengers, Forrest_Gump, John_Wick, Joker, Morbis, Pulp_Fiction, SpiderMan, The_Avengers, The_dark_night, Thor_Ragnarok]
dataset = pd.concat(dfs, ignore_index=True)

dataset.drop_duplicates()
dataset.reset_index()
# dataset.drop('index', axis=0, inplace=True)
dataset
```

## Creating ID's for each reviewer

```python
dataset['user_id'] = pd.factorize(dataset['User'])[0] + 1
dataset
```

- The reason for differences between the number of rows and the number of ID's is that some users left comments more than once

```python
duplicates = dataset[dataset.duplicated(subset='User')]
duplicates
```

## Cleaning Date Feature

```python
dataset[['Day', 'Month', 'Year']] = dataset['Date of Review'].str.split(' ', expand=True)
dataset = dataset.drop('Date of Review', axis=1)
dataset
```

## Combining Reviews

```python
dataset['Review'] = dataset['Review Title'] + ".\n" + dataset['Review']

dataset.drop(columns=['Review Title'], inplace=True)
dataset
```

## Calculating the number of Tokens for each review

```python
def count_tokens(review):
    if isinstance(review, str):
        tokens = word_tokenize(review)
        return len(tokens)
    else:
        return 0

dataset['TokenCount'] = dataset['Review'].apply(count_tokens)
dataset
```

## Make an unique ID for each REVIEW

```python
dataset['ReviewID'] = pd.factorize(dataset['Review'])[0] + 1
dataset
```

## Final Dataset

```python
final = dataset[['MovieID','user_id','User', 'Total Votes', "User's Rating out of 10", 'ReviewID', 'Review', 'TokenCount', 'Day', 'Month', 'Year']]
final
```

```python
final.dropna(inplace=True)
final.info()
```

## Replace Errors With Median

```python
final['User\'s Rating out of 10'] = pd.to_numeric(final['User\'s Rating out of 10'], errors='coerce')

temp = final['User\'s Rating out of 10'].median()

final.loc[final['User\'s Rating out of 10'].isna(), 'User\'s Rating out of 10'] 
```

```python
final['User\'s Rating out of 10'].fillna(temp, inplace=True)
final["User's Rating out of 10"].unique()
```

```python
final.to_csv('MovieReview.csv', index=False)
```
